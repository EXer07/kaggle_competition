```python
import pandas as pd 
```

```python
train = pd.read_csv(r'F:\Thai_speech\kaggle_competition\HackOWASP ML Contest\train_ml.csv',index_col='id')
test = pd.read_csv(r'F:\Thai_speech\kaggle_competition\HackOWASP ML Contest\test_ml.csv',index_col='id')
```

```python
mapping = {f'f{i}': f'feature_{i}' for i in range(1, 101)}
test_renamed = test.rename(columns=mapping)
test_renamed
```

#### Correlation Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr_matrix = train.corr()
corr_with_target = corr_matrix['target'].drop('target')
selected_corr = corr_with_target[abs(corr_with_target) > 0.03]
df_train_selected = train[selected_corr.index.tolist() + ['target']]
```

```python
print(df_train_selected.isnull().sum())
```

```python
# เติมด้วยค่าเฉลี่ยของแต่ละ column
df_train_filled = df_train_selected.fillna(df_train_selected.mean())
df_train_filled
```

```python
target_column = 'target'
feature_columns = [col for col in df_train_selected.columns if col != target_column]
df_test_selected = test_renamed[feature_columns]
df_test_filled = df_test_selected.fillna(df_test_selected.mean())
df_test_filled
```

#### Model Development

```python
X_train = df_train_filled.drop(columns=['target'])
y_train = df_train_filled['target']
X_test = df_test_filled
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=X_test.columns)
X_test_scaled_df
```

```python
from sklearn.neural_network import MLPRegressor
mlp = MLPRegressor(hidden_layer_sizes=(256, 128, 64), activation='relu',
                   solver='adam', learning_rate_init=0.001, max_iter=1000, random_state=42)
mlp.fit(X_train_scaled, y_train)
```

```python
y_pred = mlp.predict(X_test_scaled)
```

```python
submission = pd.DataFrame({
    "id": X_test_scaled_df.index,
    "target": y_pred
})

submission.to_csv("submission.csv", index=False)
```
