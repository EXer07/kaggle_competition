```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors, rdMolDescriptors, Lipinski
```

```python
train = pd.read_csv('train.csv',index_col='Batch_ID')
train
```

```python
test = pd.read_csv('test.csv',index_col='Batch_ID')
test
```

```python
def smile_extraction(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return None   
    return {
        "MolWt": Descriptors.MolWt(mol),
        "TPSA": Descriptors.TPSA(mol),
        "NumHDonors": Descriptors.NumHDonors(mol),
        "NumHAcceptors": Descriptors.NumHAcceptors(mol),
        "MolLogP": Descriptors.MolLogP(mol),
        "HeavyAtomCount": rdMolDescriptors.CalcNumHeavyAtoms(mol),
        "NumRotatableBonds": Lipinski.NumRotatableBonds(mol),
        "NumAliphaticRings": rdMolDescriptors.CalcNumAliphaticRings(mol),
        "NumAromaticRings": rdMolDescriptors.CalcNumAromaticRings(mol),
        "FractionCSP3": rdMolDescriptors.CalcFractionCSP3(mol),
    }
feature_smiles = train['Smiles'].apply(smile_extraction)
```

```python
train_feature_smiles = train['Smiles'].apply(smile_extraction)
test_feature_smiles = test['Smiles'].apply(smile_extraction)

df_train_smile = pd.DataFrame(train_feature_smiles.tolist())
df_test_smile = pd.DataFrame(test_feature_smiles.tolist())

df_train_smile.index = train.index
df_test_smile.index = test.index

df_concat_train = pd.concat([train.drop(columns=['Smiles']),df_train_smile],axis=1)
df_concat_test = pd.concat([test.drop(columns=['Smiles']),df_test_smile],axis=1)
```

```python
def select_features_by_corr(df, target_name='T80', threshold=0.5):
    
    corr_matrix = df.corr()

    corr_with_target = corr_matrix[target_name].drop(target_name)

    selected_features = corr_with_target[(corr_with_target > 0.5) | (corr_with_target < -0.5)]
    df_selected = df[selected_features.index.tolist() + ['T80']]
    

    return df_selected 
```

```python
df_concat_train_full = select_features_by_corr(df_concat_train)
df_concat_test_full = df_concat_test[df_concat_train_full.columns]
```

```python
df_concat_train_full.to_csv('preprocess_train.csv')
```

```python
df_concat_test_full.to_csv('preprocess_test.csv')
```

#### Model Development

```python
X_train = df_concat_train_full.drop(columns=['T80'])
y_train = df_concat_train_full['T80']
```

```python
X_train
```

```python
X_test = df_concat_test_full.copy()
```

```python
X_test  = X_test.drop(columns=['T80'])
X_test 
```

```python
import xgboost as xgb

model = xgb.XGBRegressor(
    objective='reg:squarederror',
    n_estimators=100,
    learning_rate=0.1,
    max_depth=4,
    random_state=42
)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
submission = pd.DataFrame({
    "Batch_ID": X_test.index,
    "T80": y_pred
})

submission.to_csv("submission.csv", index=False)
```
