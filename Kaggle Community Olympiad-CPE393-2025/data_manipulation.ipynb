```python
import pandas as pd 
import numpy as np
```

```python
df_train = pd.read_csv('train.csv')
```

```python
df_train.isnull().sum()
```

```python
# deal with Hepatomegaly
print(df_train['Hepatomegaly'].value_counts()) # the number of these samples in each class are very close to each other . then drop feature
df_train['Hepatomegaly'].isnull().sum()
df_train = df_train.drop(columns=['Hepatomegaly'],axis=1)
df_train
```

```python
df_train
```

```python
def fill_with_mode(df,col_list):
    for column in col_list:
        mode_val = df[col_list].mode().iloc[0]
        df[col_list] = df[col_list].fillna(mode_val)
    return df
col_mode_list = ['Drug','Ascites','Spiders','Cholesterol']
df_train = fill_with_mode(df_train,col_mode_list)
```

```python
# Copper           6707
# Alk_Phos         6621
# SGOT             6624
# Tryglicerides    8556
# Platelets         558
# Prothrombin        18

df_train[['Copper','Alk_Phos','Tryglicerides','Platelets','Prothrombin']]

def fill_with_mean(df,column_list):
    for column in column_list:
        mean_value = df[column].mean()
        df[column] = df[column].fillna(mean_value)
    return df

columns_to_fill = ['Copper', 'Alk_Phos', 'Tryglicerides', 'Platelets', 'Prothrombin','SGOT']

df_train = fill_with_mean(df_train, columns_to_fill)
df_train
# df_train.drop(columns=['Hepatomegaly'],axis=1)
```
