#### Correlation Analysis

```python
def label_encode(df,column):
    map_dict = {}
    for col in column:
        unique_col = df[col].unique()
        mapping = {value:index for index,value in enumerate(unique_col)}
        df[col] = df[col].map(mapping)
        map_dict[col] = mapping
    return df , map_dict
col_list = ['Drug','Sex','Ascites','Spiders','Edema']
train_encoded , map_dict = label_encode(df_train,col_list)
test_encoded , map_dict_test = label_encode(df_test,col_list)
```

```python
# covert label into numeric
df_train['Status'] = df_train['Status'].map({"D":0,"CL":1,"C":2})
```

```python
# features selection with Correlation Analysis
corr_matrix = df_train.corr()
print("Correlation Matrix:")
print(corr_matrix)

# ตรวจสอบฟีเจอร์ที่มีความสัมพันธ์สูงกับ 'Status'
status_correlation = corr_matrix['Status']
print("\nCorrelation with Status:")
print(status_correlation)


important_features = status_correlation[abs(status_correlation) > 0.2].index.tolist()
print("\nImportant Features with high correlation to Status:")
print(important_features)
```

```python
from sklearn.ensemble import RandomForestClassifier

X = df_train.drop(columns=['Status'])
y = df_train['Status']

model = RandomForestClassifier()
model.fit(X, y)

# Feature Importance
feature_importance = model.feature_importances_

# build dataframe
importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importance
})

# sorting
importance_df = importance_df.sort_values(by='Importance', ascending=False)


print("Feature Importance (sorted):")
print(importance_df)
```

```python
selected  = importance_df[importance_df['Importance'] > 0.05]
selected
```
